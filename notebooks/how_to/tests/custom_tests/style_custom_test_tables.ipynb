### Key concepts

**Model documentation**: A structured and detailed record pertaining to a model, encompassing key components such as its underlying assumptions, methodologies, data sources, inputs, performance metrics, evaluations, limitations, and intended uses. It serves to ensure transparency, adherence to regulatory requirements, and a clear understanding of potential risks associated with the model's application.

**Documentation template**: Functions as a test suite and lays out the structure of model documentation, segmented into various sections and sub-sections. Documentation templates define the structure of your model documentation, specifying the tests that should be run, and how the results should be displayed.

**Tests**: A function contained in the ValidMind Library, designed to run a specific quantitative test on the dataset or model. Tests are the building blocks of ValidMind, used to evaluate and document models and datasets, and can be run individually or as part of a suite defined by your model documentation template.

**Custom tests**: Custom tests are functions that you define to evaluate your model, dataset, or another model artifact. These functions can be registered with the ValidMind Library and run with `run_test()`.

**Test results**: Outputs generated by ValidMind tests. Test results can include descriptions, tables, figures, metrics, and raw data, and can be logged to the ValidMind Platform.

## Style a table with native pandas

The simplest way to color a custom test table is to return a native pandas `Styler`. This lets you use the styling logic your team may already have in pandas while preserving common cell-level styles in ValidMind output.

ValidMind preserves these pandas cell styles when displaying or logging the table:

- `background` and `background-color`
- `color`
- `font-weight`
- `text-align`

The example below styles rows by review status, highlights the observed values column, and formats numeric values as percentages.

````python
@vm.test("my_custom_tests.PandasStyledTable")
def pandas_styled_table():
    df = pd.DataFrame(
        [
            {
                "Check": "Missing values",
                "Observed": 0.008,
                "Threshold": "<= 1.0%",
                "Status": "Pass",
            },
            {
                "Check": "Outlier rate",
                "Observed": 0.027,
                "Threshold": "<= 2.0%",
                "Status": "Review",
            },
            {
                "Check": "Schema match",
                "Observed": "1 column missing",
                "Threshold": "0 missing",
                "Status": "Fail",
            },
        ]
    )

    row_backgrounds = {
        "Pass": "#F6FBF8",
        "Review": "#FFF9E8",
        "Fail": "#FDF0F2",
    }
    status_styles = {
        "Pass": "background-color: #DFF6E8; color: #0B6B3A; font-weight: 600; text-align: center",
        "Review": "background-color: #FFF1BF; color: #835B00; font-weight: 600; text-align: center",
        "Fail": "background-color: #FFD8DF; color: #A0182D; font-weight: 600; text-align: center",
    }

    def style_row(row):
        row_background = row_backgrounds[row["Status"]]
        styles = []

        for column in row.index:
            if column == "Observed":
                styles.append("background-color: #EAF4FF; color: #083E44; text-align: center")
            elif column in ["Check", "Threshold"]:
                styles.append(f"background-color: {row_background}; color: #083E44; font-weight: 600")
            else:
                styles.append("")

        return styles

    return (
        df.style.format({"Observed": lambda value: f"{value:.1%}" if isinstance(value, float) else value})
        .apply(style_row, axis=1)
        .map(lambda value: status_styles.get(value, ""), subset=["Status"])
    )
````

## Use ValidMind structured cells

Native pandas styling is the preferred starting point for full-cell table colors. You can also return structured cell dictionaries directly when your custom test builds table output without pandas `Styler`.

Return a dictionary with `value` plus style keys such as `bgcolor`, `color`, `fontWeight`, and `textAlign` for cells that need color.

````python
def styled_cell(
    value,
    background,
    color="#083E44",
    font_weight=400,
    text_align="left",
):
    return {
        "value": value,
        "bgcolor": background,
        "color": color,
        "fontWeight": font_weight,
        "textAlign": text_align,
    }


@vm.test("my_custom_tests.StructuredStyledTable")
def structured_styled_table():
    rows = [
        {
            "check": "Missing values",
            "observed": "0.8%",
            "threshold": "<= 1.0%",
            "status": "Pass",
            "row_background": "#F6FBF8",
            "status_background": "#DFF6E8",
            "status_color": "#0B6B3A",
        },
        {
            "check": "Outlier rate",
            "observed": "2.7%",
            "threshold": "<= 2.0%",
            "status": "Review",
            "row_background": "#FFF9E8",
            "status_background": "#FFF1BF",
            "status_color": "#835B00",
        },
        {
            "check": "Schema match",
            "observed": "1 column missing",
            "threshold": "0 missing",
            "status": "Fail",
            "row_background": "#FDF0F2",
            "status_background": "#FFD8DF",
            "status_color": "#A0182D",
        },
    ]

    return pd.DataFrame(
        [
            {
                "Check": styled_cell(row["check"], row["row_background"], font_weight=600),
                "Observed": styled_cell(row["observed"], "#EAF4FF", text_align="center"),
                "Threshold": styled_cell(row["threshold"], row["row_background"], font_weight=600),
                "Status": styled_cell(
                    row["status"],
                    row["status_background"],
                    color=row["status_color"],
                    font_weight=600,
                    text_align="center",
                ),
            }
            for row in rows
        ]
    )
````
